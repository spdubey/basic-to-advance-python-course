```python
import numpy, matplotlib.pyplot as plt
x = numpy.arange(5)
plt.plot(x, x, label='linear')
plt.plot(x, x*x, label='square')
plt.plot(x, x*x*x, label='cube')
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Polynomial Graph')
plt.legend(loc = 'center left')
#'upper center', 'lower center', 'center left', 'center right'
plt.show()
```

## Working with multiple plot figures

```python
for i in [1,2]:
    plt.figure(i)
    print("***canvas number {}********".format(i))
    for x in range(1,10):
        plt.subplot(3,3,x) # here x = rows, y = cols, x = figure number in the plot canvas
        plt.grid()
        plt.plot([1,2,3])
    plt.show()
```

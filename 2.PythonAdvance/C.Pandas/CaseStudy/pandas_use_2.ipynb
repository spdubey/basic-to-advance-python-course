### <span style="color:Blue">16) Treat missing value</span>.

```python
pd.isnull(data).sum()/len(data) *100
```

```python
data['Unit Price'] = data['Unit Price'].fillna(-9999999)
```

```python
pd.isnull(data).sum()
```

```python
# Let us fill missing value with mode and mean depending on data type except 'Total Profit' -
# -coz we will deal with it seperately
data.dtypes
```

```python
for col in data.columns:
    print("--",col)
    if col == 'Total Cost':
        continue
    else:
        if data.dtypes[col] == 'object':
            data[col] = data[col].fillna(data[col].mode()[0])
        elif data.dtypes[col] == 'float64':
            data[col] = data[col].fillna(data[col].mean())
```

```python
pd.isnull(data).sum()
```
